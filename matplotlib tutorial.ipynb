```python
k=np.linspace(-1,1,1000)
y=np.sin(2*np.pi*k)
plt.plot(k,y)
```

```python
k=np.linspace(0,10,1000)
y=np.sin(2*np.pi*k)
plt.plot(k,y)
```

```python
#parabola
i=np.linspace(10,-10,30)
j=i**2
plt.plot(j,i)
```

```python
l=np.linspace(-5,5,3000)
km=np.cos(2*np.pi*l)
plt.plot(l,km*np.exp(-l))
```

```python
#dampened oscillations
i=np.linspace(-5,5,10000)
j=np.exp(-i)
k=np.cos(2*np.pi*i)
plt.plot(i,j*k)
```

```python
fig=plt.figure()
plt.subplot(221)#rows cols the index of box
p=plt.plot(i,k,'b--')
plt.subplot(222)#rows cols the index of box
q=plt.plot(i,j,'g--')
plt.subplot(212)#rows cols the index of box
r=plt.plot(i,k*j,'k')
```

```python
t=np.arange(0,10,0.1)
x=np.sin(t)
y=np.cos(t)
plt.plot(t,x,'r.',
         t,y,'b.',
         t,x*y,'g')
```

```python
t=np.arange(0,10,0.1)
x=np.sin(t)
y=np.cos(t)
plt.plot(t,x,'r.',
         t,y,'b.',
         t,x+y,'g')
```
