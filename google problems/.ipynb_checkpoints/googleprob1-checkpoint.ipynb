```python
from itertools import combinations as com
from itertools import permutations as per
import numpy as np
```

```python
n=int(input())
```

```python
while n>0:
    x,y=map(int,input().split())
    l=np.arange(x)
    comb=com(l,y)
    o=0
    for i in list(comb): 
        print("comb:",i)
        perm=per(i,len(i))
        for j in list(perm):
            print("perm:",j)
            o=o+1
    n-=1
    
print("o:",o)
pe=per([3,2,1,0],4)
#for i in list(pe):
print("final:",len(list(pe)))
```
