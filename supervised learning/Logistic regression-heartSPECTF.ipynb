```python
dtest=pd.read_csv('https://archive.ics.uci.edu/ml/machine-learning-databases/spect/SPECTF.test',header=None)
dtest.columns = ['OVERALL_DIAGNOSIS','F1R','F1S','F2R','F2S','F3R','F3S','F4R','F4S','F5R','F5S','F6R','F6S','F7R','F7S',
                  'F8R','F8S','F9R','F9S','F10R','F10S','F11R','F11S','F12R','F12S','F13R','F13S','F14R',
                  'F14S','F15R','F15S',
                  'F16R','F16S','F17R','F17S','F18R','F18S','F19R','F19S','F20R','F20S','F21R','F21S','F22R','F22S']
df1=pd.DataFrame(dtest)
Xtest=df1['F20S']#presently considering only one feature impacting the target
Ytest=df1['OVERALL_DIAGNOSIS']
xtest=np.array(Xtest)
ytest=np.array(Ytest)
x=(x-x.min())/(x.max()-x.min())
xtest=(xtest-xtest.min())/(xtest.max()-xtest.min())
```

```python
def hypothesis(a,x,b):
    return(1/(1+np.exp(-(a*x+b))))
```

```python
def error(a,x,b,y):
    e=0
    m=len(x)
    for i in range(m):
        e+=-(y[i]*np.log(hypothesis(a,x[i],b)))-((1-y[i])*np.log(1-hypothesis(a,x[i],b)))
    return (1/m)*e
```

```python
def step_gradient(a,x,b,y,learning_rate):
    grad_a=0
    grad_b=0
    m=len(x)
    for i in range(m):
        grad_a+=(hypothesis(a,x[i],b)-y[i])*x[i]
        grad_b+=(hypothesis(a,x[i],b)-y[i])
    a=a-(grad_a*learning_rate)
    b=b-(grad_b*learning_rate)
    return a,b
```

```python
def descend(init_a,x,init_b,y,learning_rate,iterations):
    a=init_a
    b=init_b
    for i in range(iterations):
        e=error(a,x,b,y)
        if i%1000==0:
            print(f"error: {e}, a:{a}, b:{b}")
        a,b=step_gradient(a,x,b,y,learning_rate)
    return a,b
```

```python
init_a=0
init_b=0
iterations=10000
learning_rate=0.01
```

```python
final_a,final_b=descend(init_a,x,init_b,y,learning_rate,iterations)
```

```python
def accuracy(a,x,b,y):
    correct=0
    m=len(x)
    for i in range(m):
        pred=hypothesis(a,x[i],b)
        if pred>0.5:
            z=1
        else:
            z=0
        if y[i]==z:
            correct+=1
    print(f"Accuracy: {correct/len(y)}")
```
