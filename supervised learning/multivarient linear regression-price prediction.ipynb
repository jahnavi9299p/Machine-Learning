```python
#LSTAT fits better for a polynomial curve
#RM fits better for a linear curve
x=np.array(X)
y=np.array(Y)
z=np.array(Z)
```

```python
x_train,x_test,y_train,y_test,z_train,z_test=train_test_split(x,y,z,test_size=0.2)#0.2 is 20% of the data in dataset
```

```python
def hypothesis(a,b,c,x,z):
    return((a*x)+(b*z)+c)
```

```python
def error(a,b,c,x,z,y):
    e=0
    m=len(x)
    for i in range(m):
        e+=np.power((hypothesis(a,b,c,x[i],z[i])-y[i]),2)
    return (1/(2*m))*e
```

```python
def step_gradient(a,b,c,x,z,y,learning_rate):
    grad_a=0
    grad_b=0
    grad_c=0
    m=len(x)
    for i in range(m):#is the summation i.e.j(a,b)
        grad_a+=(1/m)*(hypothesis(a,b,c,x[i],z[i])-y[i])*x[i]#represents the gradient eqn i.e. j(a,b) differentiated wrt a 
        grad_b+=(1/m)*(hypothesis(a,b,c,x[i],z[i])-y[i])*z[i]#represents the gradient eqn i.e. j(a,b) differentiated wrt b
        grad_c+=(1/m)*(hypothesis(a,b,c,x[i],z[i]))
    a=a-grad_a*learning_rate
    b=b-grad_b*learning_rate
    c=c-grad_c*learning_rate
    return a,b,c
```

```python
def descent(init_a,init_b,init_c,x,z,y,learning_rate,iterations):
    a=init_a
    b=init_b
    c=init_c
    for i in range(iterations):
        e=error(a,b,c,x,z,y)
        if i%1000==0:
            print("Error: {}, a: {}, b: {}, c: {}".format(np.sqrt(e),a,b,c))
        a,b,c=step_gradient(a,b,c,x,z,y,learning_rate)
    return a,b,c
```

```python
a=0
b=0
c=0
learning_rate=0.005
iterations=10000
```

```python
final_a,final_b,final_c=descent(a,b,c,x_train,z_train,y_train,learning_rate,iterations)
```
