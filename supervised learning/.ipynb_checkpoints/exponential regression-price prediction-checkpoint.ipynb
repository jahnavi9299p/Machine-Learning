```python
#LSTAT fits better for a polynomial curve
#RM fits better for a linear curve
x=np.array(Z)
y=np.array(Y)
```

```python
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.2)#0.2 is 20% of the data in dataset
```

```python
len(x_test)
```

```python
def hypothesis(a,b,x):
    return(a*np.exp(b*x))
```

```python
def error(a,b,x,y):
    e=0
    m=len(x)
    for i in range(m):
        e+=np.power((hypothesis(a,b,x[i])-y[i]),2)
    return (1/(2*m))*e
```

```python
def step_gradient(a,b,x,y,learning_rate):
    grad_a=0
    grad_b=0
    m=len(x)
    for i in range(m):#is the summation i.e.j(a,b)
        grad_a+=(1/m)*(hypothesis(a,b,x[i])-y[i])*np.exp(b*x[i])#represents the gradient eqn i.e. j(a,b) differentiated wrt a 
        grad_b+=(1/m)*(hypothesis(a,b,x[i])-y[i])*a*x[i]*np.exp(b*x[i])#represents the gradient eqn i.e. j(a,b) differentiated wrt b
    a=a-grad_a*learning_rate
    b=b-grad_b*learning_rate
    return a,b
```

```python
def descent(init_a,init_b,x,y,learning_rate,iterations):
    a=init_a
    b=init_b
    for i in range(iterations):
        e=error(a,b,x,y)
        if i%1000==0:
            print("Error: {}, a: {}, b: {}".format(np.sqrt(e),a,b))
        a,b=step_gradient(a,b,x,y,learning_rate)
    return a,b
```

```python
a=0.6013812669291706
b=-1.45966837393826
learning_rate=0.005
iterations=10000
```

```python
final_a,final_b=descent(a,b,x_train,y_train,learning_rate,iterations)
```

```python
plt.plot(x_train,y_train,'r.',
         x_train,hypothesis(a,b,x_train),'b.',
         x_train,hypothesis(final_a,final_b,x_train),'g.')
```
