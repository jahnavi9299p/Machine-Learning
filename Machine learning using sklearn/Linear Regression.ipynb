```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.model_selection import  train_test_split
from sklearn.datasets import load_boston
```

```python
boston=load_boston()
```

```python
boston.data
```

```python
print(boston.DESCR)
```

```python
features=pd.DataFrame(boston.data,columns=boston.feature_names)
features
```

```python
target=pd.DataFrame(boston.target,columns=["TARGET"])#features and target are in different dataframes
target
```

```python
df=pd.concat([features,target],axis=1)
df
```

```python
correlation=df.corr('pearson')#pearson gives linear correlation-when u increase 1 unit of feature and how does it respond
correlation#below is correlation matrix
```

```python
X=np.array(df['RM'])
Y=np.array(df['TARGET'])
```

```python
from sklearn.preprocessing import MinMaxScaler
```

```python
scaler=MinMaxScaler()
```

```python
X=X.reshape(-1,1)
Y=Y.reshape(-1,1)
print(X.shape)
print(Y.shape)
```

```python
x=scaler.fit_transform(X)
y=scaler.fit_transform(Y)
```

```python
plt.plot(x,y,'r.')
```

```python
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.3)
```

```python
from sklearn.linear_model import LinearRegression
```

```python
regressor=LinearRegression()
```

```python
regressor.fit(x_train,y_train)
```

```python
y_pred=regressor.predict(x_test)
```

```python
y_pred
```

```python
for i in zip(y_pred,y_test):
    print(i)
```

```python
from sklearn.metrics import mean_squared_error,r2_score
```

```python
mean_squared_error(y_test,y_pred)
```

```python
plt.scatter(x_train,y_train,color='red')
plt.plot(x_train,regressor.predict(x_train),'b')
plt.xlabel("RM")
plt.ylabel("TARGET")
plt.title("Linear Regression(Train)")
```

```python
plt.scatter(x_test,y_test,color='red')
plt.plot(x_test,regressor.predict(x_test),'b')
plt.xlabel("RM")
plt.ylabel("TARGET")
plt.title("Linear Regression(Train)")
```

```python
r2_score(y_test,y_pred)#tells the best possible fit:percentage of points to be fitted
```
