```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.model_selection import  train_test_split
```

```python
ffire=pd.read_csv('C:\\Users\\Janu&Vaibh\\Desktop\\HDSfoundation\\Datasets\\forestfires.csv')
```

```python
forestfires=pd.DataFrame(ffire)
```

```python
forestfires
```

```python
features=pd.DataFrame(forestfires,columns=['temp','RH','wind','rain'])
features
```

```python
target=pd.DataFrame(forestfires,columns=["area"])#features and target are in different dataframes
target
```

```python
df=pd.concat([features,target],axis=1)
df
```

```python
correlation=df.corr('pearson')#pearson gives linear correlation-when u increase 1 unit of feature and how does it respond
correlation#below is correlation matrix
```

```python
X=np.array([df['temp'],df['RH'],df['wind'],df['rain']])
Y=np.array(df['area'])
```

```python
from sklearn.preprocessing import MinMaxScaler
```

```python
scaler=MinMaxScaler()
```

```python
X=np.transpose(X)
Y=Y.reshape(-1,1)
print(X.shape)
print(Y.shape)
```

```python

x=scaler.fit_transform(X)
y=scaler.fit_transform(Y)
```

```python
#plt.plot(x,y,'r.')
```

```python
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.3)
```

```python
from sklearn.linear_model import LinearRegression
```

```python
regressor=LinearRegression()
```

```python
regressor.fit(x_train,y_train)
```

```python
y_pred=regressor.predict(x_test)
```

```python
y_pred
```

```python
for i in zip(y_pred,y_test):
    print(i)
```

```python
from sklearn.metrics import mean_squared_error,r2_score
```

```python
mean_squared_error(y_test,y_pred)
```

```python
r2_score(y_test,y_pred)#tells the best possible fit:percentage of points to be fitted
```
